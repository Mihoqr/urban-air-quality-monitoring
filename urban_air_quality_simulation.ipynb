```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Number of simulated IoT readings
num_records = 100

# Data list for urban air quality monitoring
data = []

for _ in range(num_records):
    record = {
        "timestamp": datetime.now() - timedelta(minutes=np.random.randint(0, 1440)),  # Last 24 hours
        "station_id": f"Station_{np.random.choice(['A', 'B', 'C'])}",
        "PM2.5": round(np.random.uniform(5, 200), 1),  # PM2.5 µg/m³
        "CO2": round(np.random.uniform(300, 600), 1),  # CO2 ppm
        "temperature": round(np.random.uniform(15.0, 35.0), 1),  # Celsius
        "humidity": round(np.random.uniform(20.0, 80.0), 1),  # %
        "AQI": np.random.randint(0, 300)  # Simulated AQI index
    }
    data.append(record)

# Convert to DataFrame
df = pd.DataFrame(data)

# Save to files
df.to_csv("urban_air_quality_data.csv", index=False)
df.to_json("urban_air_quality_data.json", orient="records")

# Display first rows
df.head()
```
